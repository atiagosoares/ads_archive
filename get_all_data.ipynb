```python
def get_page_results(page_id, outfile):

    params = {
        'ad_reached_countries' : 'CA',
        'search_page_ids' : page_id,
        'ad_delivery_date_min' : '2021-05-01',
        'ad_delivery_date_max' : '2021-09-21',
        'fields' : ','.join(FIELDS),
        'access_token' : API_KEY
    }

    full_data = list()
    url = "https://graph.facebook.com/v12.0/ads_archive"
    trials = 0
    next = True
    page_count = 0
    
    while next:
        try:
            # Get page data
            if trials == 10:
                raise
            
            r = requests.get(url, params)
            current_url = r.url

            # Extend full data with current data
            data = r.json()['data']
            full_data.extend(data)

            # get next page
            params['after'] = r.json()['paging']['cursors'].get('after')
            if not params['after']:
                next = False

            #Save previous request url for debbuging purposes
            paging_previous = r.json()['paging']
            trials = 0
            page_count += 1
        
        except KeyError as e:
            print('Key Error. Waiting and retrying')
            print(e)
            time.sleep(1)
            trials += 1
        
        except requests.exceptions.ConnectionError as e:
            print('ConectionError. Waiting and retrying')
            print(e)
            time.sleep(1)
            trials += 1
        
        except Exception as e:
            break
    
    # Save current data to file
    with open(outfile, 'w') as file:
        json.dump(full_data, file)
```
