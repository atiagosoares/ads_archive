```python
import os
import json
import pandas as pd
```

```python
files = os.listdir('data')
```

```python
j = list()
for file in files:
    with open('data/' + file, 'r') as f:
        file_data = json.load(f)
        j.extend(file_data)
```

```python
demographic_distribution = list()
delivery_by_region = list()

for idx, entry in enumerate(j):
    try:
        dbr = entry['delivery_by_region']
        dbr_df = pd.DataFrame.from_dict(dbr)
        dbr_df['ad_id'] = entry['id']
        delivery_by_region.append(dbr_df)

        dd = entry['demographic_distribution']
        dd_df = pd.DataFrame.from_dict(dd)
        dd_df['ad_id'] = entry['id']
        demographic_distribution.append(dd_df)
    except:
        print(idx)
        print('Error')
```

```python
ads_df = pd.json_normalize(j)
```

```python
ads_df.columns = ads_df.columns.str.replace('.', '_', regex = True)
```

```python
ads_df.to_csv('data/ads.csv')
```

```python
delivery_by_region_df = pd.concat(delivery_by_region)
delivery_by_region_df.to_csv('data/delivery_by_region.csv')
```

```python
demographic_distribution_df = pd.concat(demographic_distribution)
demographic_distribution_df.to_csv('data/demographic_distribution.csv')
```
